<a name="for"></a> For Loops
===
`for` loop digunakan untuk melakukan iterasi pada suatu sequence (seperti list, tuple, set dan dictionary).

```python
fruits=['apel','mangga','pisang']
for i in fruits:
    print(i)
```

```python
fruits=('apel','mangga','pisang')
for i in fruits:
    print(i)
```

```python
fruits={'buah_1':'apel','buah_2':'mangga','buah_3':'pisang'}
for i in fruits:
    print(i)
```

```python
for i in 'apel':
    print(i)
```

<a name="range"></a> Range Function
===

```python
for i in range(6):
    print(i)
```

```python
for i in range(0,30,7):#range(start,stop,step)
    print(i)
```

<a name="forelse"></a> For Else
===

```python
for x in range(6):
    print(x)
else:
    print("Finally finished!")
```

<a name="while"></a> While Loops
===
`while` loop digunakan untuk melakukan iterasi selama kondisi yang diberikan benar.

```python
i=0
while i < 5:
    print(i)
    i += 1
```

<a name="whileelse"></a> While Else
===

```python
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is no longer less than 6")
```

<a name="break"></a> Break dan Continue
===
Pernyataan break digunakan untuk memberhentikan iterasi sebelum iterasi dilakukan ke semua item. Pernyataan continue digunakan untuk memberhentikan eksekusi langkah/kegiatan pada iterasi yang sekarang dan melanjutkan ke eksekusi langkah/kegiatan di iterasi sebelumnya (Skip Iterasi).

```python
i=5
while i < 15:
    print(i)
    if i == 8:
        break
    i += 1
```

```python
fruits=['apel','mangga','pisang']
for i in fruits:
    if i=='apel':
        continue
    print(i)
```

===

```python
for i in range(3):
    for j in range(2):
        print(i,j)
```
