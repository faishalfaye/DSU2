```python
#Pendefinisian fungsi
def nama_function(argument_1, argument_2):
    # Langkah atau kegiatan dalam fungsi
    return 'Nilai' #Untuk mengembalikan hasil yang diinginkan

# Penggunaan fungsi yang telah didefinisikan
nama_function(value_1, value_2)
```

- **Pendefinisian Fungsi**
    - Keyword `def`, sebagai penanda pada python bahwa kita akan mendefinisikan suatu fungsi.
    - Pemberian nama fungsi setelahnya.
    - Pemberian nama pada variabel input (Argument) yang akan digunakan di dalam fungsi.
    - Penggunaan `return` untuk menggembalikan suatu hasil.

- **Penggunaan function**
    - Untuk memanggil function yang telah didefinisikan dapat dilakukan dengan menulis nama function diikuti dengan parenthesis.

<a name='default'></a>Default values
---
Pada argumen input dapat dimasukkan nilai yang berperan sebagai `default values`, dimana nantinya menjadi nilai argumen input tersebut ketika pemanggilan fungsi tanpa disertai dengan nilai input pada argumen tersebut.

```python
def thank_you(name):
    print("You are doing good work, %s!"%name)
    print("Thank you very much for your participation in DSU BATCH 2.\n")
    
thank_you('Adriana')
thank_you('Billy')
thank_you()
```

```python
def thank_you(name=None):
    # This function prints a two-line personalized thank you message.
    if name:
        print("You are doing good work, %s!"%name)
        print("Thank you very much for your participation in DSU BATCH 2.\n")
    
thank_you('Eja')
thank_you('Khaerul')
thank_you()
```

<a name='position'></a>Positional Argument
---
Pada function posisi/urutan nilai yang dimasukkan berpengaruh. 

```python
def describe_person(first_name, last_name, age):
    print("First name:",first_name)
    print("Last name:",last_name)
    print("Age:",age)
    print('----------------------------')
```

```python
describe_person('Abu','Isan',17)
describe_person('Isan','Abu',17)
describe_person(17,'Isan','Abu')
```

Ketika urutan/posisi nilai yang dimasukkan berubah maka nilai akan dimasukkan pada argumen yang berbeda.

<a name='keyword'></a>Keyword Argument
---
Memasukkan nilai pada argumen juga dapat dilakukan dengan memasangkannya sesuai dengan nama argumen ketika fungsi di definisikan.

```python
describe_person(age=17,last_name='Abi',first_name='Isan')
```

<a name='latihan'></a>Latihan

Buatlah sebuah fungsi yang menerima argumen berupa :
- Sebuah list `Bilangan` berisi bilangan bulat.
- Nilai integer `K` dengan default values 0.

Dengan output berupa list yang berisi seluruh elemen dari list `Bilangan` yang bernilai tidak sama dengan `K`

```python
lis = list(map(int, input().strip().split()))
k = 0

def non_k(lis, k):
  true_list = [x for x in lis if x != k]
  return true_list
  
non_k(lis, k)
```
