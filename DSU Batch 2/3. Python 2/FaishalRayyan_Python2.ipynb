```python
import csv

tp = 0
fn = 0
fp = 0
tn = 0

with open("/content/klasifikasi.csv", "r") as csvfile:
  reader = csv.reader(csvfile)
  next(reader, None)
  for row in reader:
    if (row[0] == "1") and (row[1] == "1"):
      tp += 1
    elif (row[0] == "1") and (row[1] == "0"):
      fp += 1
    elif (row[0] == "0") and (row[1] == "0"):
      tn += 1
    elif (row[0] == "0") and (row[1] == "1"):
      fn += 1


```

```python
print(tp)
print(fp)
print(tn)
print(fn)
```

```python
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = (2 * precision * recall) / (precision + recall)
accuracy = (tp + tn) / (tp + tn + fn + fp)
specificity = tn / (tn + fp)
```

```python
print(precision)
print(recall)
print(f1)
print(accuracy)
print(specificity)
```
