Exploratory Data Analysis (EDA) adalah suatu prosedur / kegiatan menganalisa data untuk memahami karakteristik data tersebut. Seringkali pada saat terjun ke lapangan, data yang kita dapat tidak datang dalam keadaan rapi dan bersih, bahkan seringkali data yang kita peroleh sangat berantakan, diperlukan usaha ekstra untuk mempersiapkan data tersebut untuk siap dilakukan analisis. Dalam pembuatan machine learning sendiri, EDA yang tidak dilakukan dengan benar memiliki resiko terhambatnya langkah selanjutnya dalam proses pembuatan machine learning tersebut.

Tujuan EDA :
- Membuat atau menemukan hipotesis terhadap fenomena yang ditemukan pada data
- Membuat asumpsi berdasarkan inferens statistik
- Mendukung pemilihan teknik statistik yang tepat
- Memberi alasan terhadap pengambilan data lanjutan

Data Understanding: Relevance :
- Data apa yang tersedia?
- Seberapa banyak (dan lama) data tersedia?
- Ada yang memiliki label? (Variabel Target)

### Keterangan Data

Variables :
- Pregnancies : Banyaknya kehamilan
- Glucose : Konsentrasi plasma glukosa pada 2 jam saat test toleransi glukosa oral
- BloodPressure : Tekanan darah diastolik (mmHg)
- SkinThickness : Ketebalan kulit triceps (mm)
- Insulin : 2-Hour serum insulin (mu U/ml)
- BMI : Body mass index (Berat dalam kg/(tinggi dalam m)^2)
- DiabetesPedigreeFunction : Diabetes pedigree function
- Age : Umur (Tahun) 
- Outcome : Diabetes atau tidak 0 : Tidak, 1 : Diabetes

Sumber data :https://www.kaggle.com/uciml/pima-indians-diabetes-database

# Noise

Noise dapat terjadi karena:
- Kesalahan instrumen pengukuran: Misal di alat IoT pada saat cuaca buruk/baterai yang lemah.
- Kesalahan input/entry
- Transmisi yang tidak sempurna
- inkonsistensi penamaan

# Outlier

Data yang memiliki karakteristik secara signifikan berbeda dengan kebanyakan data lainnya menurut suatu kriteria tertentu yang ditetapkan.

# Missing Value

Missing Values merupakan salah satu proses dalam ‘membersihkan data’ yang mengidentifikasi dan menghandle missing value, apa itu missing value? Missing value adalah istilah untuk data yang hilang

# Penyebab Missing Value ?

Data yang hilang ini bisa disebabkan oleh beberapa hal, salah satu contohnya adalah
- Error pada data entry, baik itu human error ataupun kesalahan pada sistem
- Pada data survey, bisa disebabkan oleh responden yang lupa mengisi pertanyaan, pertanyaan yang sulit dimengerti, ataupun pertanyaan enggan diisi karena merupakan pertanyaan yang sensitif

# Lineplot

Biasa digunakan untuk menampilkan nilai numerik pada satu garis di mana x sebagai variabel kategori dan y sebagai variabel jumalh data.

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 5, 3, 5, 7, 8]
plt.figure(figsize=(5,7))#ukuran plot 5 x 7
plt.plot(x, y) 
plt.title('Judul Plot')
plt.xlabel('Keterangan Koordinat X')
plt.ylabel('Keterangan Koordinat Y')
plt.legend()
plt.show()
```

```python
sns.lineplot(x,y)
```

# Count Plot / Bar Charts

Bar charts menggunakan batang horizontal atau vertikal (bagan kolom) untuk menampilkan perbandingan numerik diskrit di seluruh kategori. Satu sumbu bagan menunjukkan kategori tertentu yang dibandingkan dan sumbu lainnya mewakili skala nilai diskrit.

```python
sns.countplot(df.Outcome)
```

# Histogram

```python
sns.histplot(data=df,x='Pregnancies',hue='Outcome')
```

# Boxplot

Boxplot biasa digunakan untuk menampilkan distribusi data secara keseluruhan melalui kuartil. Garis memanjang sejajar dari kotak dikenal sebagai "whisker(kumis)", yang digunakan untuk menunjukkan variabilitas di luar kuartil atas dan bawah. Pencilan terkadang diplot sebagai titik individu yang sejajar dengan kumis. Box Plot dapat digambar secara vertikal maupun horizontal. Meskipun Box Plot mungkin tampak primitif dibandingkan dengan histogram, mereka memiliki keuntungan mengambil lebih sedikit ruang, yang berguna ketika membandingkan distribusi antara banyak kelompok atau dataset. 

```python
sns.boxplot(data=df[['Glucose','BloodPressure']])
```

```python
def category(x):
    if x==0:
        return 'Non-Diabetes'
    else:
        return 'Diabetes'
```

```python
df.Outcome=df.Outcome.apply(lambda x:category(x))
```
